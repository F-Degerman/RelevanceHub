# 2026-02-26


## Agentic implementation

**Focus:** Using AI as active ‘agents’ in operations – not just as chat tools.

**What it means:**

* Instead of building custom AI models from scratch, use existing tools and connect them.
* AI agents can:

  * Plan tasks
  * Call tools (APIs, databases, internal systems)
  * Execute steps in a process autonomously
* It is about integrating AI into real workflows – not just experimenting.

Strategic shift: from ‘proof of concept’ to real operational use.

---



## Use tools wisely – don’t build everything yourself

**Focus:** Efficiency and time-to-market.

**What it means:**

* Companies should avoid building their own LLMs or infrastructure unless absolutely necessary.
* Instead:

  * Use established platforms
  * Combine ready-made services
  * Build on top of existing ecosystems

It is more an integration issue than a research issue.

---



## Agents – determinism vs indeterminism

**Focus:** Control vs creativity in AI systems.

**Determinism:**

* Same input → same output
* Predictable
* Important in business-critical systems

**Indeterminism:**

* Same input can produce different responses
* More creative and flexible
* Good for ideation work, but harder to quality-assure

From a management perspective:
How much variation can we accept in production systems?

## Internal infrastructure

**Focus:** AI maturity in the organization.

**What it includes:**

* Data access (APIs, databases, internal systems)
* Security and access control
* Logging and traceability
* GPU/compute resources (if relevant)
* CI/CD and version control

Many companies are limited more by their infrastructure than by AI technology.

## Git branches

**Focus:** Structure in AI and code development.

**What it means:**

* Clear version control
* Separate experiments from production
* Ability to test different AI flows in parallel
* Branch strategy for prompts and agent logic

AI development needs the same discipline as traditional software development.

## Project planning in focus – AI as process partner

**Focus:** AI before, during, and after the project.

Instead of only using AI for code or text:

* Let AI create project plans
* Risk analyses
* Milestones
* Communication structure
* Test strategies

AI becomes a co-pilot for the entire project process.

---


## Ask an LLM for a prompt plan before the project

**Focus:** Meta-level AI usage.

Instead of writing prompts directly:

* Ask the model to create a prompt strategy
* Ask for structure on how AI should be used in the project
* Ask for improvements to your own prompts

Example:

> “Create an optimal prompt structure for running an IT project with the help of AI.”

It is about systematizing AI usage.
