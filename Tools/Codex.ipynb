# Codex (OpenAI, ChatGPT+)
- som hjälp vid studier

Med Codex installerat i VS-code har jag chatbot och studiematerial samlat på ett ställe. Jag kan lägga in relevanta filer för mina projekt/repon och hänvisa till dem både vid programmering och vid teoretiska frågor. Filerna (från kursen) lägger jag i mappar som ignoreras av `.gitignore`, så att de bara finns lokalt och inte delas på GitHub.  
  
På så sätt kan jag ha allt relevant material för kursen, mina egna anteckningar och filer på samma ställe som mitt faktiska repo (som jag vill visa offentligt).   
  
Det gör att jag enkelt kan ställa frågor och vara tryggare med att få ett relevant svar utifrån kursens innehåll.

**OBS:**  
- Du kan inte hänvisa till det du skrivit tidigare i appen (ChatGPT). Jag har själv lagt upp mina chattar och filer i projekt där och chatten(Codex) kunde inte ta del av det. Rekommenderar därför att börja använda Codex i början av en kurs, så att du inte behöver flytta över filer eller starta om konversationer. 
   
För ChatGPT+ (appen) finns det en begränsning på 25 filer per projekt. Nu behöver jag inte tänka på det då filerna inte läggs in i appen, utan istället hänvisas till lokalt.
