# Scrum 

**Scrum** is one of the most common agile frameworks for organizing and carrying out development projects. It is used to create value step by step, get fast feedback, and adapt when requirements change. In the course syllabus, Scrum is mentioned as an example of an agile project method that you should be able to use and explain. 

# The Scrum Structure

## 1. Roles

### Product Owner (PO)

Responsible for **what** should be built.

* Prioritizes tasks
* Manages product requirements
* Owns the Product Backlog
* Represents the needs of the customer or the business

### Scrum Master

Responsible for **how Scrum works**.

* Helps the team follow Scrum
* Removes obstacles
* Leads meetings
* Coaches the team in agile ways of working

### Development Team

Responsible for creating the product.

* Cross-functional team
* Self-organizing
* Often consists of developers, testers, data engineers, ML engineers, and others

---

## 2. Artifacts

### Product Backlog

A prioritized list of everything that may need to be done.

Example in an AI project:

| Priority | Task             |
| -------- | ---------------- |
| 1        | Collect data     |
| 2        | Build pipeline   |
| 3        | Train model      |
| 4        | Deploy to Azure  |
| 5        | Create dashboard |

---

### Sprint Backlog

The tasks the team chooses to complete during the next sprint.

---

### Increment

The completed result after the sprint.

Examples:

* Working data pipeline
* Trained model
* API for inference
* Dashboard

---

## 3. The Scrum Cycle

```text
Product Backlog
       v
Sprint Planning
       v
Sprint (1-4 weeks)
       v
Daily Scrum
       v
Completed delivery (Increment)
       v
Sprint Review
       v
Sprint Retrospective
       v
New Sprint
```

---

## 4. Meetings (Events)

### Sprint Planning

The start of the sprint.

Questions:

* What are we going to do?
* How are we going to do it?

---

### Daily Scrum

15 minutes every day.

Each person usually answers:

1. What did I do yesterday?
2. What am I going to do today?
3. Are there any obstacles?

---

### Sprint Review

Present the result.

* Demonstration of what has been built
* Feedback from stakeholders

---

### Sprint Retrospective

Reflection on the way of working.

Questions:

* What worked well?
* What worked less well?
* What can be improved?

---

## Example from Data Engineering

Assume you are going to build an ML solution in Azure.

### Sprint 1

* Set up a Git repository
* Create a Docker environment
* Fetch data

### Sprint 2

* Build an ETL pipeline
* Data cleaning

### Sprint 3

* Train an ML model
* Evaluate results

### Sprint 4

* Deploy to Azure
* Monitoring and logging

Each sprint delivers something that actually works.

---

# Scrum in One Sentence

**Scrum is a way to organize projects where a team works in short iterations (sprints), delivers value continuously, and improves both the product and the way of working through regular feedback.**
