how **attention can be rewritten explicitly as a kernel method**, which is exactly what models like Performer exploit.

---

# 1. Start from attention

$$
\mathrm{Attention}(Q, K, V) = \mathrm{softmax}(QK^T)V
$$

For a single query $q$:

$$
\text{Att}(q) = \sum_i \frac{\exp(q \cdot k_i)}{\sum_j \exp(q \cdot k_j)} \, v_i
$$

---

# 2. Interpret softmax as a kernel

Define a **softmax kernel**:

$$
K(q, k_i) = \exp(q \cdot k_i)
$$

Then attention becomes:

$$
\text{Att}(q) = \frac{\sum_i K(q, k_i) \, v_i}{\sum_j K(q, k_j)}
$$

## Key insight

This is:

> **normalized kernel regression**

Compare with classical kernel methods:

$$
f(x) = \sum_i K(x, x_i) \, y_i
$$

Attention = kernel smoother + normalization.

---

# 3. Feature map trick (core idea in Performer)

We want to rewrite the kernel as:

$$
K(q, k) = \phi(q)^T \phi(k)
$$

If possible, then:

$$
\exp(q \cdot k) \approx \phi(q)^T \phi(k)
$$

## Then attention becomes

$$
\text{Att}(q) =
\frac{\sum_i \phi(q)^T \phi(k_i) \, v_i}{\sum_j \phi(q)^T \phi(k_j)}
$$

Factor out $\phi(q)$:

$$
\text{Att}(q) =
\frac{\phi(q)^T \left(\sum_i \phi(k_i) \, v_i\right)}{\phi(q)^T \left(\sum_j \phi(k_j)\right)}
$$

---

# 4. Why this is powerful

Now computation becomes:

- Precompute:
  - $S = \sum_i \phi(k_i) \, v_i$
  - $Z = \sum_i \phi(k_i)$
- Then for each query:
  - compute dot products with $S$ and $Z$

## Result

- Standard attention: **$O(n^2)$**
- Kernelized attention: **$O(n)$**

---

# 5. This is what Performer does

It uses **random feature maps** to approximate:

$$
\exp(q \cdot k)
$$

Example idea:

- Use random projections to build $\phi(x)$
- Approximate the softmax kernel efficiently

---

# 6. Big picture connection

| Concept | SVM / Kernel | Attention |
| --- | --- | --- |
| Kernel | $K(x, x')$ | $\exp(q \cdot k)$ |
| Feature map | $\phi(x)$ | $\phi(q), \phi(k)$ |
| Output | weighted sum | weighted sum |
| Normalization | optional | required (softmax) |

---

# Final intuition

Attention is essentially:

> **a normalized, learned kernel regression operating on tokens**

Performer shows:

> you can **turn attention into a true kernel method** using feature maps

## One-line takeaway

> Transformers secretly behave like **adaptive kernel machines**, and efficient variants make this connection explicit.
